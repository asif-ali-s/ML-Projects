### Importing the Dependencies, Loading Data and Analyzing

## Model Development

```python
X = sdata.drop(columns=60, axis=1)
Y = sdata[60]
```

```python
print(X)
```

```python
print(Y)
```

```python
#split the data into train and test datasets to measure the model generalization to predict unseen data of the model
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

### Model development with kNN

```python
neighbors = np.arange(1,14)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
```

```python
for i,k in enumerate(neighbors):
 knn = KNeighborsClassifier(n_neighbors=k)
 knn.fit(X_train, y_train)
 train_accuracy[i] = knn.score(X_train, y_train)
 test_accuracy[i] = knn.score(X_test, y_test)
```

```python
plt.title('k-NN Varying number of neighbors')
plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
plt.plot(neighbors, train_accuracy, label='Training accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```

it can be seen that accuracy for both the training as well as the testing data decreases with the increasing number of neighbors, so k=2 would be a safe number to assume.

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
knn.score(X_test,y_test)
```

```python
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

In this model, mines were predicted as mines 34 times, 1-time mines were predicted as rocks, rocks were predicted as mines 9 times, and 19 times rocks were predicted as rocks.
